```python
#colormaap and to change color after specific value
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = {"Product" : ['A','J','A','L','R','A','R','J','L','A'],
        "Country" : ['USA','Japan','Europe','Japan','Europe','USA','USA','Japan','USA','Europe'],
        'Sales':[44000,43000,54000,95600,79000,67000,68000,38000,59000,60000]
       }
df = pd.DataFrame(data)
#fig, ax = plt.subplots(figsize=(10,4))
df.groupby(['Country','Product']).sum()['Sales'].unstack().plot(kind="bar")
plt.suptitle("SALES AMONG COUNTRIES")
```
