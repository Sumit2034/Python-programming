```python
df = pd.DataFrame(lend)
```

```python
df1 = df.fillna(0)
```

```python
df1['Rank'] = df1['Rank'].replace(regex = True, to_replace = '[A-Z]', value = 0)# ranking can't be in negative it can be either 0 or above
```

```python
df1['Rank'][df1['Rank'] < 0] = 0
df1
```

```python
df1['Lender Asset Size'].unique()
```

```python
df1 = df1.drop(['Unnamed: 10'],axis = 1)
```

```python
df1
```

```python
df1 = df1.drop([0],axis=0)
```

```python
df1
```

```python
df1['TA Ratio1'] = df1['TA Ratio1'].astype('str')
a=0
count=0
for i in df1['TA Ratio1']:
    try:
        if i != '-':
            i = float(i)
            a+=i
            count+=1
    except ValueError:
        pass
print(a)
print(count)
mean_of_columns = a/count
print(mean_of_columns)
```

```python
df1["TA Ratio1"] = df1["TA Ratio1"].replace("-",mean_of_columns)
```

```python
df1["CC Amount/TA1"].describe
```

```python
df1['CC Amount/TA1'] = df1['CC Amount/TA1'].astype('str')
b=0
count1=0
for i in df1['CC Amount/TA1']:
    try:
        if i != '.':
            i = float(i)
            b+=i
            count1+=1
    except ValueError:
        pass
print(b)
print(count1)
mean_of_columns1 = b/count1
print(mean_of_columns1)
df1["CC Amount/TA1"] = df1["CC Amount/TA1"].replace("    .  ",mean_of_columns1)
```

```python
df1['TBL Ratio1'] = df1['TBL Ratio1'].astype('str')
c=0
count2=0
for i in df1['TBL Ratio1']:
    try:
        if i != '-':
            i = float(i)
            c+=i
            count2+=1
    except ValueError:
        pass
print(c)
print(count2)
mean_of_columns2 = c/count2
print(mean_of_columns2)
df1["TBL Ratio1"] = df1["TBL Ratio1"].replace("-",mean_of_columns2)
```

```python
df1['Amount ($1,000)'] = df1['Amount ($1,000)'].astype('str')
d = 0
count3 =0 
for i in df1['Amount ($1,000)']:
    try:
        if i != '-':
            i = int(i)
            d = d+i
            count3+=1
    except ValueError:
        pass
print(d)
print(count3)
mean = d/count3
print(mean)
df1['Amount ($1,000)'] = df1['Amount ($1,000)'].replace("-",mean)
```

```python
df1['Number '] = df1['Number '].astype('str')
e=0
count4=0
for i in df1['Number ']:
    try:
        if i != '-':
            i = int(i)
            e+=i
            count4+=1
    except ValueError:
        pass
print(e)
print(count4)
mean_of_columns4 = e/count4
print(mean_of_columns4)
df1["Number "] = df1["Number "].replace("-",mean_of_columns4)
```

```python
df1
```

```python
df1['Lender Asset Size'] = df1['Lender Asset Size'].replace(['>$50B               ','$10B-$50B           ','>$10B               '],[0,1,2])
```

```python
df1
```

```python
df1['Name of Lending Institution'] = df1['Name of Lending Institution'].astype('str') 
df1['HQ State'] = df1['HQ State'].astype('str') 
df1['Rank'] = df1['Rank'].astype('int') 
df1['TA Ratio1'] = df1['TA Ratio1'].astype('float') 
df1['TBL Ratio1'] = df1['TBL Ratio1'].astype('float') 
df1['Amount ($1,000)'] = df1['Amount ($1,000)'].astype('float') 
df1['Number '] = df1['Number '].astype('int') 
df1['Lender Asset Size'] = df1['Lender Asset Size'].astype('int') 
df1['CC Amount/TA1'] = df1['CC Amount/TA1'].astype('float') 
```
